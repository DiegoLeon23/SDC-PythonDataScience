```python
df_metodoanticonceptivo.columns = ['edad_esposa', 'educacion_esposa', 'educacion_esposo', 'numero_hijos_nacidos',
                   'religion_esposa', 'trabaja_esposa', 'ocupacion_esposo',
                   'ind_nivel_vida', 'expo_medios', 'metodo_anticonceptivo']
df_metodoanticonceptivo.head()
```

__2-Segmentacion de la edad de la esposa.__

```python
edad_intervalos=[15,25,30,35,40,45,50]
edad_segm = pd.cut(df_metodoanticonceptivo['edad_esposa'],edad_intervalos) 
pd.value_counts(edad_segm)
```

__3-Grafica de boxplot de la edad de la esposa.__

```python
sns.boxplot(list(df_metodoanticonceptivo['edad_esposa']))
plt.title('Boxplot de Edad de la esposa')
plt.show()
```

__4-Graficar el diagrama de dispersion y el histrograma para mostrar la relacion entre las variables edad de la esposa y número de hijos nacidos.__

```python
# Distribución de salario
df_metodoanticonceptivo['edad_segm'].hist(bins=edad_intervalos ,color='yellow', ec='black').set_title('Distribución de Edad_esposa')
plt.show();
```

```python
num_hijos=[0,2,4,6,8,10,12,14,16]
df_metodoanticonceptivo['numero_hijos_nacidos'].hist(bins=num_hijos,color='yellow', ec='black').set_title('Distribución de numero_hijos_nacidos')
plt.show();
```
