## Resumen de una red neuronal de una capa

Este capítulo trata sobre redes neuronales multicapa, sobre como trabajan y como se entrenan para resolver problemas complejos. Sin embargo, antes de que nos sumerjamos más en la arquitectura de las redes neuronales multicapa, vamos a reiterar brevemente algunos de los conceptos de las redes neuronales de una capa.   

```python
Image(filename='C:/Users/Social Data/Desktop/wbs/Redes Neuronales/images/12_01.png', width=600)
```

Para realizar las clasificaciones binarias, se implementa el algoritmo Adaline y para aprender los coeficientes ponderados del modelo, se usa el algoritmo de optimización de descenso del gradiente. En cada paso por el conjunto de entrenamiento, actualizamos el vector de peso $w$ mediante la siguiente regla de actualización. 

$$ w := w + \bigtriangleup w, where \bigtriangleup w = -\eta\bigtriangledown J (w)$$

En otras palabras, computamos el gradiente según el conjunto de entrenamiento completo y actualizamos los pesos del modelo haciendo un paso en la dirección opuesta al gradiente $\bigtriangledown J(w)$. Con el fin de encontrar los pesos óptimos del modelo, optimizamos una función objetiva que definimos acomo función de coste de la **suma de los errores cuadráticos(SSE)** $\bigtriangledown  J(w)$. Ádemas, multiplicamos el gradiente por el factor, el ragon de aprendizaje $\eta,$ que tuvimos que elegier atentamente para equilibrar la velocidad del aprendizaje contra el riesgo de superar el mínimo global de la función coste. 

En la optimizacióndel descenso del gradiente, actualizamos todos los pesos a la vez despúes de cada época, y definimos la derivada parcial para cada peso w, en el vecto de peso $w$ del modo siguiente.


$$\frac{\partial}{\partial_j}J(w) = \sum_i \left(y^{(i)}-\alpha^{(i)}\right)x^{(i)}_j$$

Aquí, $y^{(i)}$ es la etiqueta de clase de destino de una muestra en particular $x^{(i)}, y$ $a^{(i)} $ es la activación del a nuerona, una función lineal en el caso especial de Adaline. Ádemas, definimos la función de activación $\phi(.)$ del siguiente modo.   

$$\phi(z)=z=a$$

En este caso, la entrada de red z es una combinación lineal de los pesos que conectan la entrada a la capa de salida:

$$z=\Sigma_jw_jx_j = w^Tx$$

Al tiempo que utilizabamos la activación $\phi(z)$ para calcular la actualización del gradiente, implementamos una funcipon de umbral para comprimir la salida del valor continuo en etiquetas de clase binaria para la predicción.

$$
    \left.
        \begin{aligned}
            1 if g(z)\geq0\\
            -1 otherwise
        \end{aligned}
    \right\}
\quad\text{\hat{y}}
$$

Ádemas, aprendimos un truco para acelerar el aprendizaje del modelo: la denominada optimización del **descenso del gradiente estocástico**. El descenso del gradiente estocástico aproxima el coste a partir de una sola muestra de entrenamiento (aprendizaje online) o de un pequeño subconjunto de muestras de entrenamiento (aprendizaje por minilotes). Utilizaremos este concepto mas adelante en este capitulo, cuando implementamos y entrenamos un perceoton multicapa.  Aparte del aprendizajemás rapido - debido a las actualizaciones de pesos mas frecuentes en comparación con el descenso de gradiente-, su naturaleza ruidoasa ea algo que se ve como beneficioso cuando se entrenan redes neuronales multicapa con funciones de la activación no linea, las cuales no tienen una función de coste convexa. Áqui, el ruido añadido puede ayudar a escapar de los mínimos locales de coste, pero trataremos este tema con más detalle más adelante en este capítulo.

En esta sección, aprenderás cómo conectar múltiples neuronas simples a una red neuronal prealimentada multicapa; este tipo especial de red $completamente conectada$ también se denomina **Perceptron Multicapa(ML.P.**, de ingles $Multilayer Perceptron$). La siguiente imagen elustra el concepto de un MIP de tres capas.

El MLP representado en la imagen anterior tiene una capa de entrada, una capa oculta y una capa de salida. Las unidades en la capa oculta estan conectadas completamente a la capa de entrada, y la capa de salida está conectada completamente a la capa oculta. Si una red así tiene más de una capa oculta, también la denominamos **red neuronal artificial profunda**.  

Como se muestra en la figura anterior hemos presentado la unidad i de la activacion de la $l^{ra}$ capa como $a^{(l)}_i$. Para hacer que las implementaciones del código y de las mátematicas sean como mas intuitivas, no utilizaremos indices numéricos para referirnos a las capas sino el superíndice $in(input)$ para la capa de entrada, el superíndice $h (oculta)$ para la capa oculta, y en el superíndice $o(output)$ para la capa de salida. Por ejemplo, $a^{(in)}_i$ se refiere al valor $i^o$ en la capa de entrada; $a^{(h)}_i$, se refiere a la unidad $i^a$ en la capa oculta;y $a^{(out)}_i$ se refiere a la unidad $i^a$ en la capa de salida. 


En este caso, las unidades de activación $a^{(in)}_0a^{(h)}_0$ son las unidades de sesgo, que hemos configurado igual a 1. La activación de las unidades en la capa de entrada es sencillamente su entrada más la unidad de sesgo.

$$ a^{(in)}=\left[\begin{smallmatrix}
                a^{(in)}_0\\
                a^{(in)}_1\\
                .\\
                .\\
                .\\
                a^{(in)}_m
                \end{smallmatrix}\right]
= \left[\begin{smallmatrix}
                1\\
                x^{(in)}_1\\
                .\\
                .\\
                .\\
                x^{(in)}_m
                \end{smallmatrix}\right]
 $$


Cada undidad de la capa $l$ está conectada < todas las unidades de la capa $l + 1$ mediante un coeficiente ponderado. Por ejemplo, la conexión entre la unidad $k^a$ en la capa $l$ y la unidad $j^a$ en la capa $l+1$ desde escribirse como $w^{(l+1)}_{k,j}$. Volviendo a la imagen anterior, hemos representado la matriz ponderada que conecta la capa de entrada a la de salida como $W^{(h)}$, y hemos escrito la matriz que conecta la capa oculta a la de salidad como $W^{(out)}$. 

Mientras que una unidad en la capa de salida bastaría para una tarea de clasificación  binaria, en la imagen anterior hemos visto una forma más general de red neuronal, que nos permite realizar clasificaciones multiclase mediante una generalización de la técnica $One-versus-All(OvA)$. ## Activar una red neuronal mediante la propagación hacia delante

En esta sección, describiremos el proceso de **propagación hacia delante** para calcular la salida de un modelo MLP. Para entender cómo se ajusta al contexto del aprendizaje de un modelo MLP, vamos a resumir el procedimiento de aprendizaje MLP en tres simples pasos:
<br>

- Empezando  en la capa de entrada, propagamos hacia delante los patrones de los datos de entrenamiento a través de la red para generar una salida.
- Basándonos en la salida de la red, calculamos el error que queremos minimizar mediante una función de coste que describiremos más adelante.
- Propagamos hacia adelante el error, encontramos su derivada con respecto a cada peso en la red y actualizamos el modelo.

Por último, después de repetir estos tres pasos durante múltiples épocas y aprender los pasos del MLP, utilizamos la propagación hacia delante para calcular la salidad de la red y aplicar una función de umbral que obtenga las etiquetas de las calses predicahs en una representación $one-hot$, que hemos descrito en la sección anterior.

Ahora, veamos los pasos de la propagación hacia delante por separado para generar una salida a partir de los patrones en los datos de entrenamiento. Como cada unidad en la capa oculta está conectada a todas las unidades en las capas de entrada, primero calculamos la unidad de activación de la capa oculta $a^{(k)}_i$ de este modo:

$$z^{(h)}_1=a^{(in)}_0w^{(h)}_{0.1}+a^{(in)}_1w^{(h)}_{1.1}+...+a^{(in)}_mw^{(h)}_{m.1}$$

$$a^{(h)}_1=\phi(z^{(h)}_{i})$$

Aquí, $z^{(k)}_l$ es la entrada de red y $\phi(.)$ es la función de activación, que debe ser diferenciable para aprender los pesos que conectan las neuronas mediante un método basado en el gradiente. Para ser capaces de resolver problemas complejos como una clasificación de imagenes, necesitamos, necesitamos funciones de activación no lienales en nuestro modelo MLP; por ejemplo, la función de la activación sigmoide (logística) que recordarás de la sección sobre la regresión logística. 

$$\phi(z)=\frac{1}{1+e^{-z}}$$

Como recordamos, la función sigmoide es una curva con forma de S que mapea la entrada en red Z dentro de la distrbución logística en el rango 0 a 1, que corta el eje $y$ donde $z$ = 0, como se muestra en la siguiente imagen.

El MLP es un ejemplo típico de una red neuronal artificial prealimentada. El termino **prealimentada** se refiere al hecho de que cada capa sirve de entrada a la capa siguiente sin bucles, en comparación con las redes neuronales recurrentes -aarquitectura que trataremos más adelante -. El termino Perceptrón Multicapa puede sonar un poco confuso, pues las neurona artificiales en estas arquitecturas de redes dson habitualmente unidades sigmoides, no perceptrones. De forma intuitiva, podemos pensar en las neuronas del MLP como unidades de regresión logística que devuelve valores en los rangos continuos entre 0 y 1.
<br>

Por motivos de eficacia y lectura del código, ahora escrbiremos la activación de un modo mas compacto mediante el concepto del álgebra lineal básica, que nos permitirá vectorizar nuestra implementación del código mediante NumPy en lugar de escribir múltiples bucles for de Python anidados y computacionales costosos:


$$z^{(h)}=a^{(in)}W^{(h)}$$

$$a^{(h)}=\phi\left(z^{(h)}\right)$$

Áqui, **$a^{(in)}$** es nuestro vector de características dimensional $1$ x $m$ de una muestra $x^{(in)}$ más de una unidad de sesgo. $w^{(h)}$ es una matriz ponderada dimensional $m$ x $d$ es el número de unidades en la capa oculta. Después de reakizar la multiplicación matriz-vector, obtenemos el vector de entrada de red dimensional $1$ x $d$ **$z^{(h)}$** para calcular la activación $a^{(h)}$ (donde $a^{(h)} \in \Re^{1xd})$. Ádemas, podemos generalizar este calculo en todas las n muestras del conjunto de entrenamiento.

$$Z^{(h)}= A^{(in)}W^{(h)}$$

Áqui, $A^{(in)}$ es ahora de una matriz $n$ x $m$, y la multiplicación matriz-matriz dará como resultado una amtriz de entrada de red dimensional $n$ $d$ $Z^{(h)}$. Por último, aplicamos la función de activación $\phi(.)$ a cada valor en la matriz de entrada de red para obtener la matriz de activación $n$ $x$ $d$ $A^{(h)}$ para la siguiente capa(aquí la cpa de saliad). 

$$A^{(out)}=\phi\left(Z^{(h)}\right)$$

De forma similar, podemos escribir la activación de la capa de salida en formato vectorizado para múltiples muestras.

$$Z^{(out)}=A^{(h)}W^{(out)}$$

Aquí, multiplicamos la matriz $d$ $x$ $t$ $W^{(out)}$ (t es el número de unidades de salida) por la matriz dimensional $n$ $x$ $t$ $Z^{(out)}$ (las columnas en esta matriz representan las salidas para cada muestra).
<br>

Por Último, aplicamos la función de activación de sigmoide para obetener la salida de valor continuo de nuestra red:

$$A^{(out)}=\phi\left(Z^{(out)}\right), A^{(out)} \in \Re^{nxt}$$
