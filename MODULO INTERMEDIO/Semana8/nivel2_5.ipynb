## Reglas de Asociación

- Las reglas de asociación son útiles para descubrir relaciones que sean de interés en conjunto de datos.
<br>

- Estas relaciones pueden ser descritas en la forma de reglas de asociación.
<br>

- Las reglas de asociación usualmente parten de una base de datos de transacciones, D.
<br>

- Cada transacción consiste de un identificador y un conjunto de items (p. ej. productos, páginas web, cursos) ${i_1,i_2, ..., i_n}$ seleccionados de todos los ítems posibles (I)

Por ejemplo, la siguiente tabla es un ejemplo de una base de transacciones:
<BR>
    
$$\begin{array}{crl}
TID & Items \\
\text{1} & \text{Bread, Milk} \\
\text{2} & \text{Bread, Diaper, Beer, Eggs} \\
\text{3} & \text{Milk, Diaper, Beer, Coke} \\
\text{4} & \text{Bread, Milk, Diaper, Beer} \\
\text{5} & \text{Bread, Milk, Diaper, Coke}
\end{array}$$
<BR>
    
- Una regla de asociación es un una implicancia de la forma $X \Longrightarrow Y$ tal que $ X \subset I, Y \subset I y X \cap Y = \emptyset$
<br>

- $X$ es conocidad como la regla antecedente, mientras que $Y$ como la regla consecuente.
<br>

- De la tabla anterior, la siguiente regla e asociación puede ser encontrada: $\{Diaper\} \longrightarrow \{Beer\}$

- Esta regla sugiere que existe una fuerte asociación entre la venta de pañales y de cerveza porque muchos clientes que compran pañales también compran cerveza. 
<br>

- Es importante notar que las reglas de asociación son de naturaleza estocástica, lo cual implica que no deben ser interpretados como verdades universales y son caracterizadas por medidas estadísticas que cuantifica  la fuerza de la asociación.
<br>

- Las reglas miden asociaciones y no deben ser interpretados de forma cuasal.

## Confianza y Soporte

- **Itemset:** Es la colección de uno o más ítems.
<br>

    $$\{Milk, Bread, Diaper\}$$
<br>

- **Soporte:** Es la frecuencia con la que ocurre un itemset.
<br>

    $$s(\{Milk, Bread, Diaper\}) = \frac{2}{5}$$
<br>

- **Regla de Asociación:** es una relación entre dos itemset.
<br>

    $$X \longrightarrow Y$$
<br>
    representa el patrón de que cuando $X$ ocurre también ocurre $Y$
    <BR>
    $$\{Milk, Diaper\} \longrightarrow \{Beer\}$$

**Medidas de Evaluación de una Regla** $X \longrightarrow Y$
<br>

- **Soporte:**
<br>

    $$s(X\longrightarrow Y) = s (X \cup Y) = \frac{ \text{# de trans. que contienen a } (X \cup Y )}{\text{# total de transacciones}}$$  
<br>

- **Confianza:**
<br>

$$c(X \longrightarrow Y) = P(X|Y) = \frac{s(X \cup Y )}{s(X)}$$

## Reglas de Asociación

- Dado un conjunto de datos, el objtivo es encontrar todas las reglas de asociación que cumplan:
<br>

- Soporte $\geq$ minsup
<br>

- Confianza $\geq$ mincon

- Generalmente se siguen 2 pasos:
<br>

    1. Generación de itemsets frecuentes: Encontrar todos los itemsets que cumplan con Soporte $\geq$ minsup.
<br>

    2. Derivación de reglas de asociación: Generar reglas de asociación a partir de los itemset frecuentes encontrados que cumplan Confianza $\geq$ mincon.

- Algoritmo A priori.
<br>

- Si un itemser es frecuente, entonces todos sus subconjuntos deben ser frecuentes también.
