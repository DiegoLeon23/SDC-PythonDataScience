```python
#Importando librerias necesarias
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder      #Para codificar una data transaccional
from mlxtend.frequent_patterns import apriori             #Para calcular los soportes de los posibles itemsets
from mlxtend.frequent_patterns import association_rules   #Para generar las reglas de asociaciones
```

### Caso1: Productos comestibles

```python
#Generamos una data de prueba
dataset = [['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
           ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']]
```

```python
#Creamos una instancia de clase TransactionEncoder
te=TransactionEncoder()
```

```python
#Aprendemos y transformar los datos
mitransaccional=te.fit(dataset)
mitransaccional=te.transform(dataset)
print(mitransaccional)
```

```python
#Metodo auxiliar
#Aprendemos y transformar los datos
mitransaccional=te.fit_transform(dataset)
print(mitransaccional)
```

```python
#Generamos las etiquetas de las columnas
print(te.columns_)
```

```python
#Generamos un dataframe
df=pd.DataFrame(mitransaccional,columns=te.columns_)
```

```python
df.head()
```

```python
#Generamos los valores de S(X) para todas las combinaciones posibles de itemsets
suport_itemsets=apriori(df, # mi dataframe
                       min_support=0.6, # soporte minimo deseado en la regla antecedente
                       use_colnames=True) # uso de las etiquetas
suport_itemsets
```

```python
#Generando las posibles reglas de asociación con un criterio de confianza
reglas=association_rules(suport_itemsets, #soporte de itemsets
                        metric='confidence', #metrica a trabajar
                        min_threshold=0.7) #umbral minimo de confianza
reglas
```

```python
#Creamos una nueva variable (de conteo) de los items que conforman el itemset de la regla antecedente
reglas['antecedentes_freq']=reglas["antecedents"].apply(lambda x: len(x))
reglas
```

```python
#Mostrar las reglas funakes de Seleccion (segun criterio del investigador o analista)
reglas[(reglas.antecedentes_freq>=2)&(reglas.lift>1.2)]
```

### Caso2: Canciones Alejandro Sanz

```python
dataset2 = [["amiga mia", "corazón partido", "y si fuera ella"],
           ["corazón partido", "cuando nadie me ve", "aprendíz"],
           ["corazón partido", "mi soledad y yo", "amiga mia"],
           ["corazón partido", "y si fuera ella", "aprendíz"],
           ["mi soledad y yo", "cuando nadie me ve", "amiga mia"],
           ["mi soledad y yo", "la fuerza del corazón", "y si fuera ella"],
           ["aprendíz", "amiga mia", "y si fuera ella"],
           ["corazón partido", "sólo vives por él", "lo sé"],
           ["corazón partido", "mi soledad y yo", "y si fuera ella"],
           ["pisando fuerte", "no es lo mismo", "cuando nadie me ve"]]
```

```python
#Creamos una instancia de clase TransactionEncoder
te2=TransactionEncoder()
```

```python
#Aprendemos y transformar los datos
mitransaccional2=te2.fit_transform(dataset2)
```

```python
#Generamos las etiquetas de las columnas
print(te2.columns_)
```

```python
#Creamos el dataframe
df2=pd.DataFrame(mitransaccional2,columns=te2.columns_)
df2.head()
```

```python
#Generamos los soportes de las posibles de las posibles reglas antecedentes con un umbral minimo
suport_itemsets2=apriori(df2, # mi dataframe
                       min_support=0.2, # soporte minimo deseado en la regla antecedente
                       use_colnames=True) # uso de las etiquetas
suport_itemsets2
```

```python
#Generando las posibles reglas de asociación con un criterio de confianza
reglas2=association_rules(suport_itemsets2, #soporte de itemsets
                        metric='lift', #metrica a trabajar
                        min_threshold=0.8) #umbral minimo de confianza
reglas2
```

```python
#Mostrar las reglas funakes de Seleccion (segun criterio del investigador o analista)
reglas2[(reglas2.confidence>=0.6)&(reglas2.lift>1.2)]
```
