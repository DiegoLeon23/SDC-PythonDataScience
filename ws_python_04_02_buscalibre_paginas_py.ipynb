```python
import time
nro_pag=16 #El numero de paginas en total
n=0 #Comienza con 0 el numero de paginas
totalx=0 #El numero total de libros
```

```python
#Definicion del dataframe donde se almacenara los libros
books_all = {'nombre':[],
 'precio_anterior':[],
 'precio_posterior':[]}
df_books_all = pd.DataFrame(books_all)
```

```python
bandera=True #El valor de condicion
while bandera:
    n=n+1
    
    url = "https://www.buscalibre.pe/libros/search?q=Big+data&page="+str(n)
   
    #Abrimos la pag del url que especificamos
    driver.get(url)
    
    time.sleep(2)
    
    #Nombres
    div_nombres = driver.find_elements(By.XPATH, "//*[@id='content']/div[3]/div/a/div[2]")
    list_new_nombres = list(map(lambda x: x.text, div_nombres))

   
    #Precio Anterior
    div_precioant = driver.find_elements(By.XPATH, "//*[@id='content']/div[3]/div/a/div[6]/div[2]/h5/del")
    list_new_precioant = list(map(lambda x: x.text, div_precioant))

    #Precio Posterior
    div_preciopst = driver.find_elements(By.XPATH, "//*[@id='content']/div[3]/div/a/div[6]/div[2]/h3")
    list_new_preciopst = list(map(lambda x: x.text, div_preciopst))
    
    #Calculadora de productos
    number_list = len(list_new_nombres)
    #Total hasta ahora
    totalx=totalx+number_list
    
    #Generar un dataframe
    books_pag = {'nombre':list_new_nombres,
     'precio_anterior':list_new_precioant,
     'precio_posterior':list_new_preciopst}
    df_books_pag = pd.DataFrame(books_pag)
    
    #Concatenar  axis=0 es las filas y 1 sería columnas
    df_books_all=pd.concat([df_books_all, df_books_pag], axis=0)

    print("La pagina "+str(n)+"° contiene "+str(number_list)+" el total hasta ahora es "+str(totalx))
    
    if n==nro_pag:
        bandera=False
```
